# Numpy Exercises

Finding the combination of two arrays using meshgrid. numpy.meshgrid is used to 'create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing.'

```python
arr1 = np.array([1, 2, 6])
arr2 = np.array([4, 6, 4])
arrn = np.array(np.meshgrid(arr1, arr2)).T.reshape(-1, 2) #meshgrid returns a list; Using -1 in reshape ensures that the 
#other dimension is automatically calculated
print(arrn)
```

Find common values between two numpy arrays using numpy.intersect1d

```python
rng = np.random.Generator(np.random.MT19937())
arr1 = rng.integers(100, 100000, 10000)
arr2 = rng.integers(100, 100000, 10000)

#Find the average intersections of the above arrays after 1000 tries
count = 0
for _ in range(1000):
    count += len(np.intersect1d(arr1, arr2))
    
print(count/1000)
```

### 2. Given a two dimensional array, how to extract unique rows?



```python
#Getting unique rows
#initial soln
#Two arrays are similar if they have the same elements in the same order

arr1 = rng.integers(0, 10, 100).reshape(-1, 5)

#arr1 == arr1[x] provides a boolean array
#Sampling all across the horizontal axis summarizes to true iff all are true
#Use nonzero() to only give the true values
#If the size of the resultant array is >1 it repeats
#Hence:

for row in arr1:
    x = (arr1 == row).all(axis=1).nonzero()
    #x is a tuple with an ndarray, x[0], showing position row
    #x[0] is what we want
    
    if len(x[0]) == 1: #does not repeat
        print(row)
```

```python
#Given solution
Z = np.random.randint(0,2,(6,3))
T = np.ascontiguousarray(Z).view(np.dtype((np.void, Z.dtype.itemsize * Z.shape[1])))
_, idx = np.unique(T, return_index=True)
uZ = Z[idx]
print(uZ)
```

```python
#Testing both solutions:
#My soln
def find_unique1(arr1):
    for row in arr1:
        x = (arr1 == row).all(axis=1).nonzero()
        if len(x[0]) == 1: #does not repeat
            print(row)
 
#Other proposed soln
def find_unique2(Z):
    uZ = np.unique(Z, axis=0)
    print(uZ)
    
arrx = np.random.randint(0, 2, (10,3))
print(arrx)
print()
print('-----find unique 1------')
find_unique1(arrx)
print()
print('-----find unique 2------')
find_unique2(arrx)

#The two functions act differently:
#find_unique1 returns non-repeating(unique) rows only
#find_unique2 returns all rows, with each row --repeating or non-repeating--
#appearing once, essentially removes repetition.
```

```python
#Hence,the final solution is:
n = 5 #the largest set we want
arr = np.arange(10000000)
np.random.shuffle(arr)
print( '----mysoln----')
print((np.sort(arr))[arr.size - n:])

#proposed soln
#slow
print( '----proposed slow----')
print(arr[np.argsort(arr)[-n:]])

#Fast
print( '----proposed fast----')
print(arr[np.argpartition(-arr, n)[:n]])

#Slow and mysoln are comparatively slow
#Fast is quite fast
#Sort returns a sorted copy whereas argsort() returns the indices of the elements
#if the array were sorted

#The fast algorithm uses partationing, partitoning the array based on index
#np.partition(ndarray, index) returns an array in which the index given will have
#its proper element (i.e -- after partitioning the element at arr[index] is the same as after sorting)
#To the left of index, are elements less than... to the right elements greater than index... although
#not guaranteed to be in their right positions!
```

### Given a sorted array C that corresponds to a bincount, how to produce an array A such that np.bincount(A) == C? 

```python
#My soln
#np.bincount(A) returns an array of size(A.max())
#each bin in Bin is no. of times Bin[index] has occured

#Create array C that is a bincount, min=0, max=1499

C = rng.integers(0, 1499, 1500)
```

```python
#C[n] for any 0 <= n < 1500 means that n appears C[n] times in C
#To build A, put as many n in A as C[n]
#size of A is sum of C
C.sort()
A = np.zeros(C.sum(), dtype = int)
pos_ina = 0
for n in range(len(C)):
    #At every n, make C[n] n's in A
    
    for _ in range(C[n]):
        A[pos_ina] = n
        pos_ina += 1
        
(np.bincount(A) == C).all()
```

```python
#The proposed soln
#uses np.repeat
A_beta = np.repeat(np.arange(len(C)), C)
(np.bincount(A_beta) == C).all()

#Overly less cumbersome
```
